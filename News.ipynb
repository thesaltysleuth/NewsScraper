```
import nltk
from newspaper import Article
```

```
#Function to call later
def processing(article):
  #NLP
  article.download()
  article.parse()
  nltk.download('punkt')
  article.nlp()
  return article
```

```
#ENTER TOPIC NAME HERE
topic='mla'


#ENTER SITE NAME HERE

######################################################################################################


#News Site
site='https://www.ndtv.com/topic/{}'.format(topic)
```

```
site
```

```
from urllib.request import Request, urlopen

req = Request(site, headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read().decode('utf-8')
```

```
webpage
```

```
import re

pattern=r"<a\s+href=\"(https:\/\/www\.ndtv\.com\/.{3,6}news\/([^\"]*))\""
x=re.findall(pattern,webpage)
```

```
#ENTER THE MAXIMUM NUMBER OF ARTICLES YOU WANT TO VIEW

n=50

#-----------------------------------------------------------------
```

```
l=len(x)
m=min(l,n)
```

```
for i in range(m):
  article=Article(x[i][0])
  article=processing(article)
  print("-----------------------------------------------------------------")
  print("")
  print(article.title)
  print("")
  print("Author:  ",article.authors)
  print("Publish date:  ",article.publish_date)
  print("Image:  ",article.top_image)
  print(article.summary)
  print("")
  print("-----------------------------------------------------------------")
```
